## Corrigiendo el tipo de dato de cada columna

```python
#Para la columna Population (2020)
#df2['Population (2020)'] = df2['Population (2020)'].str.split(',').str.join('')
df2['Population (2020)'] = df2['Population (2020)'].astype(np.int64)
df2.dtypes
```

```python
#Para la columna Yearly Change
df2['Yearly Change'] = df2['Yearly Change'].str.extract(r'(-?\d+.?\d+) %')
df2['Yearly Change'] = df2['Yearly Change'].astype(np.float64)
df2.dtypes
```

```python
#Para la columna Net Change
df2['Net Change'] = df2['Net Change'].str.split(',').str.join('')
df2['Net Change'] = df2['Net Change'].astype(np.int64)
df2.dtypes
```

```python
#Para la columna Density (P/Km²)
df2['Density (P/Km²)'] = df2['Density (P/Km²)'].str.split(',').str.join('')
df2['Density (P/Km²)'] = df2['Density (P/Km²)'].astype(np.int64)
df2.dtypes
```

```python
#Para la columna Land Area (Km²)
df2['Land Area (Km²)'] = df2['Land Area (Km²)'].str.split(',').str.join('')
df2['Land Area (Km²)'] = df2['Land Area (Km²)'].astype(np.int64)
df2.dtypes
```

```python
#Para la columna Migrants (net)
df2['Migrants (net)'] = df2['Migrants (net)'].str.split(',').str.join('')
#Comprobar que dato son los que no aparecen 
df2['Migrants (net)'].values
#Reemplazar ' ' por np.nan
df2['Migrants (net)'] = df2['Migrants (net)'].replace([' '], np.nan)
df2['Migrants (net)'].values
#Con np.nan ya puedo usar los metodos de pandas y convertir el tipo de dato a float
df2['Migrants (net)'] = df2['Migrants (net)'].astype(np.float64)
df2.dtypes
```

```python
#Para la columna Fert. Rate
df2['Fert. Rate'] = df2['Fert. Rate'].str.split('.').str.join('')
#Reemplazar 'NA' (generado por el split-join) por np.nan
df2['Fert. Rate'] = df2['Fert. Rate'].replace(['NA'], np.nan)
df2['Fert. Rate'] = df2['Fert. Rate'].astype(np.float64)
df2.dtypes
```

```python
#Para la columna Med. Age
#Reemplazar 'N.A.' por np.nan
df2['Med. Age'] = df2['Med. Age'].replace(['N.A.'], np.nan)
df2['Med. Age'] = df2['Med. Age'].astype(np.float64)
df2.dtypes
```

```python
#Para la columna Urban Pop %
df2['Urban Pop %'] = df2['Urban Pop %'].str.extract(r'(\d+) %')
df2['Urban Pop %'] = df2['Urban Pop %'].astype(np.float64)
df2.dtypes
```

```python
#Para la columna World Share
df2['World Share'] = df2['World Share'].str.extract(r'(\d+.\d+) %')
df2['World Share'] = df2['World Share'].astype(np.float64)
df2.dtypes
```

# Trabajando con df1

```python
df1.info()
```

```python
pop1 = df1.apply(lambda row: (row['area']*row['Density']),axis=1)
pop1
#Más acertado con el resultado
```

```python
columns_df_per_year = ['name']
for column in df1.columns:
    year = re.search(r".*\d{4}", column)
    if year is not None:
        columns_df_per_year.append(column[year.start():year.end()])
    else: continue
columns_df_per_year
```

```python
df1.columns
```

```python
pop_per_year = df1.loc[:, columns_df_per_year]
pop_per_year
```

```python
pop_world = pop_per_year.melt(id_vars=['name'], var_name='year', value_name='population')
pop_world
```

```python
pop_world.info() 
```

```python
pop_world['year'] = pop_world['year'].str.extract(r".*(\d{4})")
pop_world['year'] = pop_world['year'].astype(np.int64)
pop_world.dtypes
```

```python
pop_world.sample(10)
```

```python
#La población está en miles, entonces transformamos por grupo
pop_world = pd.concat([pop_world.iloc[:, 0:2], 
           pop_world.groupby('year')['population'].transform(lambda x : x*1000)], 
          axis=1)
pop_world
```

```python
#Haciendo grupos por año (year)
years = pop_world.groupby('year')
years
```

```python
years.get_group(2020)
```
